```python
#extract 2 columns - short name of player, wage in euro

player_salary = fifa[['short_name','wage_eur']]
player_salary.head()
```

```python
#highest paid player
#sorting the sub data frame - player_salary
player_salary = player_salary.sort_values(by=['wage_eur'],ascending=False)
player_salary.head()
```

```python
plt.figure(figsize=(8, 5))

#Barplot for top 5 players in terms of salary in euros. 
#Using a dataframe and converting them into list for x and y axis. 
#No use of count_values because we have already inserted it in an object.
#The values don't need to be derived from the main object.

plt.bar(list(player_salary['short_name'])[0:5],list(player_salary['wage_eur'])[0:5], color=["Blue","Green","Red","Pink","Orange"])
```
